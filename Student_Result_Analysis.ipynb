```python
#to show gender distribution weidth ,hight=10,4 
#palette for color

plt.figure(figsize=(5,6))
ax=sns.countplot(data=df,x="Gender",hue='Gender',palette='Set1')
#to print exact value
# Add labels to each bar in both containers
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
#now lets determine the parent education on student result
```

```python
gb=df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':"mean",'WritingScore':'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.show()
```
